### Identifikasi pencilan dengan IQR

```python
import numpy as np
from numpy import percentile
isNumeric =df_south.select_dtypes(include=np.number)
# print(isNumeric)
for i in list(isNumeric):
    q25, q75 = percentile(df_south[i], 25), percentile(df_south[i], 75)
    iqr = q75 - q25
    print('Percentiles: 25th=%.3f, 75th=%.3f, IQR=%.3f' % (q25, q75, iqr))
    # calculate the outlier cutoff
    cut_off = iqr * 1.5
    lower, upper = q25 - cut_off, q75 + cut_off
    # identify outliers
#     for x in range(len(df_south_pd[i])):
#         if df_south_pd[i][x] < lower or df_south_pd[i][x] > upper:
#             df_south_pd[i][x] = np.nan
    outliers = [x for x in df_south[i] if x < lower or x > upper]

    print('Identified outliers: %d' % len(outliers))
    # remove outliers
    outliers_removed = [x for x in df_south[i] if x >= lower and x <= upper]
    print('Non-outlier observations: %d' % len(outliers_removed))
df_south
```

### Menghitung presentasi nilai unik di setiap kolom

```python
for i in list(df_south):
	num = len(pd.unique(df_south[i]))
	percentage = float(num) / df_south.shape[0] * 100
	print(i,' %d, %.1f%%' % (num, percentage))
```

### Menghapus kolom dengan hanya satu nilai unik

```python
counts = df_south.nunique()
# record columns to delete
to_del = [i for i,v in enumerate(counts) if v == 1]
print(to_del)
# drop useless columns
print(df_south.shape)
for i in to_del: 
    df_south.drop(df_south.columns[i], axis=1, inplace=True)
    print(df_south.shape)
```

### Mengubah data outlayer RADIACAO GLOBAL (Kj/m²) menjadi nan

```python
from numpy import percentile
import numpy as np
q25, q75 = percentile(df_south['RADIACAO GLOBAL (Kj/m²)'], 25), percentile(df_south['RADIACAO GLOBAL (Kj/m²)'], 75)
iqr = q75 - q25
print('Percentiles: 25th=%.3f, 75th=%.3f, IQR=%.3f' % (q25, q75, iqr))
# calculate the outlier cutoff
cut_off = iqr * 1.5
lower, upper = q25 - cut_off, q75 + cut_off
for x in range(len(df_south['RADIACAO GLOBAL (Kj/m²)'])):
    if df_south['RADIACAO GLOBAL (Kj/m²)'][x] < lower or df_south['RADIACAO GLOBAL (Kj/m²)'][x] > upper:
        df_south['RADIACAO GLOBAL (Kj/m²)'][x] = np.nan
```

### Mengubah data -9999.0 menjadi Nan

```python
import numpy as np
isNumeric =df_south.select_dtypes(include=np.number)
# print(isNumeric)
for i in list(isNumeric):
    df_south[i] = df_south[i].replace(-9999.0,np.nan)
df_south
```

### Imputasi Iteratif Mean

```python
from sklearn.impute import SimpleImputer
isNumeric =df_south.select_dtypes(include=np.number)
imputer_mean = SimpleImputer(strategy='mean')
# fit on the dataset
imputer_mean.fit(isNumeric)
# transform the dataset
Xtrans_south_mean = imputer_mean.transform(isNumeric)
Xtrans_south_mean
```

```python
columns = list(df_south.select_dtypes(include=np.number).columns)
df_south_mean = df_south.copy()
for i in range(len(columns)):
    df_south_mean[columns[i]]=Xtrans_south_mean[:,i]
df_south_mean
```
